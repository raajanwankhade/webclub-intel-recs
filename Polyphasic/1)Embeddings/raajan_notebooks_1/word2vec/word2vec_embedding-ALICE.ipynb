```python
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk
from gensim.models import Word2Vec
import re
nltk.download('punkt')
```

```python
text_cricket = [
    "The Cricket World Cup is a major event in the world of sports.\n"+
    "Cricket players from different countries participate in the World Cup.\n"+
    "The final match of the World Cup was an exciting game.\n"+
    "Batsmen scored heavily in the tournament.\n"+
    "The World Cup trophy is awarded to the winning team.\n"
]
```

```python
with open('alice.txt', 'r') as file:
    text_alice = file.read()
```

```python
text_cricket
```

```python
text_alice
```

```python
def preprocess_text(text):
    # removing special characters and converting to lowercase
    proc_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    proc_text = text.lower()

    return proc_text
```

```python
# combining the text into a single string
text_cricket = ' '.join(text_cricket)
text_cricket
```

```python
# will need individual tokens of text_cricket later
cric_sents = sent_tokenize(text_cricket)
cric_sents
```

```python
cric_sents = [preprocess_text(sentence) for sentence in cric_sents]
cric_sents
```

```python
cric_sents = [word_tokenize(sentence) for sentence in cric_sents]
cric_sents
```

```python
cric_words = [token for sentence in cric_sents for token in sentence]
print(len(cric_words))
# getting unique words/tokens
cric_words = list(set(cric_words))
print(len(cric_words))
```

```python
text = text_cricket + text_alice
text
```

```python
sents = sent_tokenize(text)
sents
```

```python
sents = [preprocess_text(sentence) for sentence in sents]
sents
```

```python
sents = [word_tokenize(sentence) for sentence in sents]
sents
```

```python
# training the cbow_model
cbow_model = Word2Vec(sents,vector_size=100,window=7,sg=0,min_count=1)
# training the skip gram model
sg_model = Word2Vec(sents,vector_size=100,window=7,sg=1,min_count=1)
```

```python
cbow_model.wv["cricket"] #printing an example
```

```python
sg_model.wv.index_to_key
```

```python
cbow_df = pd.DataFrame({'word':cbow_model.wv.index_to_key,'embeddings':cbow_model.wv.vectors.tolist()})
cbow_df
```

```python
cbow_df_filtered = cbow_df[cbow_df["word"].isin(cric_words)]
cbow_df_filtered
```

```python
sg_df = pd.DataFrame({'word': sg_model.wv.index_to_key, 'embeddings': sg_model.wv.vectors.tolist()})
sg_df
```

```python
sg_df_filtered = sg_df[sg_df["word"].isin(cric_words)]
sg_df_filtered
```

```python
# Save to CSV
cbow_df_filtered.to_csv('cbow_embeddings_alice.csv', index=False)
sg_df_filtered.to_csv('skipgram_embeddings_alice.csv', index=False)
```
