```python
from transformers import BertTokenizer, BertModel
import torch
import pandas as pd
```

```python
# general pre processing (not required as we will use bert-base-uncased)
# def preproc(text,tokeniser):
#     # tokenisation
#     tokens = tokeniser.tokenize(text)

#     # lowercase
#     tokens = [tokens.lower() for token in tokens]

#     # returning the preproc_text
#     preproc_text = " ".join(tokens)
#     return preproc_text
```

```python
text = [
    "The Cricket World Cup is a major event in the world of sports.\n" +
    "Cricket players from different countries participate in the World Cup.\n" +
    "The final match of the World Cup was an exciting game.\n" +
    "Batsmen scored heavily in the tournament.\n" +
    "The World Cup trophy is awarded to the winning team.\n"
]
```

```python
# loading the BERT model here
model_name = "bert-base-uncased"
tokeniser = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
# tokenising the text
tokenised_text = tokeniser(text,return_tensors="pt",padding=True,truncation=True)
tokenised_text
```

```python
# forward pass through the model to get the embeddings
with torch.no_grad():
    outputs = model(**tokenised_text)
```

```python
# extracting the embeddings
embeddings = outputs.last_hidden_state[0]
embeddings
```

```python
embeddings = embeddings.tolist()
embeddings
```

```python
tokens = tokenised_text['input_ids'][0]
tokens
```

```python
tokens = [tokeniser.decode([token]) for token in tokens]
tokens
```

```python
df = pd.DataFrame({'word':tokens,'embeddings':embeddings})
df
```

```python
df.to_csv('word_embeddings_bert.csv', index=False)
```
