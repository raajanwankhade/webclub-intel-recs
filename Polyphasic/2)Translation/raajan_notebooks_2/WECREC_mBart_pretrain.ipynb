```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv("nlp_intel_train.csv")
df
```

```python
df_test = pd.read_csv("nlp_intel_test.csv")
df_test
```

```python
english_to_french_data = df[["en", "fr"]]
french_to_english_data = df[["fr", "en"]]
```

```python
def translate(text, src_lang, tgt_lang):
    # Set the source and target languages for the tokenizer
    tokenizer.src_lang = src_lang
    tokenizer.tgt_lang = tgt_lang

    # Encode the input text
    encoded_text = tokenizer(text, return_tensors="pt")

    # Generate the translation
    generated_tokens = model.generate(**encoded_text, forced_bos_token_id=tokenizer.lang_code_to_id[tgt_lang])

    # Decode the generated translation and skip special tokens
    translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

    return translation
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
english_sentences = english_to_french_data["en"]
french_sentences = english_to_french_data["fr"]
```

```python
index_to_translate = 900

src_lang = "en_XX"
tgt_lang = "fr_XX"
english_sentence = english_sentences.loc[index_to_translate]
translated_sentence = translate(english_sentence, src_lang, tgt_lang)

# Display the translation result
print("English Sentence:", english_sentence)
print("Translated Sentence:", translated_sentence)
```
